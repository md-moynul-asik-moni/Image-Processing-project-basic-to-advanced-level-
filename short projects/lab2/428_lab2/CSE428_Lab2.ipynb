```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, transform,color, exposure, img_as_float
```

```python
np.random.seed(3000)
```

```python
I = np.random.rand(4,3)
print(I)
```

```python
io.imshow(I)
plt.axis('off')
```

```python
I1=I+0.5
io.imshow(I1)
```

```python
I1[I1>1] = 1

io.imshow(I1)
```

```python
I2 = np.clip(I1,0,1)
io.imshow(I2)
```

```python
I3 = np.clip((I-0.3),0,1)
io.imshow(I3)
```

```python
fig,axes = plt.subplots(1, 3, figsize=(6,6))
axes[0].imshow(I,cmap="gray")
axes[0].axis("off")
axes[0].set_title('Original')

axes[1].imshow(I1,cmap="gray")
axes[1].axis("off")
axes[1].set_title('Addition by 0.5')


axes[2].imshow(I2,cmap="gray")
axes[2].axis("off")
axes[2].set_title('Subtraction by 0.3')
```

```python
I4 = np.clip(I*2,0,1)
I5 = np.clip(I/2,0,1)
```

```python
print(I4)
```

```python
plt.subplot(1, 3, 1)
io.imshow(I)
plt.axis("off")
plt.title('Original Image')


plt.subplot(1, 3, 2)
io.imshow(I4)
plt.axis("off")
plt.title('Multiplication by 2')

plt.subplot(1, 3, 3)
io.imshow(I5)
plt.axis("off")
plt.title('Division by 2')
```

```python

I = io.imread('landsat_down.jpeg')
I = img_as_float(I)

io.imshow(I)
```

```python
hist, nbins = exposure.histogram(I, nbins=256, normalize = False)

plt.plot(nbins,hist)
plt.title('Original histogram')
```

```python
I_stretched = np.clip(4*(I-0.1),0,1)
io.imshow(I_stretched )

```

```python
hist, nbins = exposure.histogram(I_stretched , nbins=256, normalize = False)
plt.plot(nbins[1:],hist[1:])
plt.title('Stretched histogram')
```
