```python
import numpy as np
from skimage import io, color, exposure, img_as_float
import matplotlib.pyplot as plt
```

# Task 1

```python
# Import the "good.jpeg" image
I = io.imread('good.jpeg')

# normalize the image
I_norm = img_as_float(I)

# perform arithmetic operations (contrast increase/decrease) on the image
I_add = np.clip(I_norm+0.5,0,1) # increase by 0.5
I_sub = np.clip(I_norm-0.4,0,1) # decrease by 0.4
I_mul = np.clip(I_norm*2,0,1)# multiply by 2
I_div = np.clip(I_norm/3,0,1) # divide by 3





```

```python

plt.figure(figsize=(15,15))
plt.subplot(1, 5, 1)
io.imshow(I_norm)
plt.axis("off")
plt.title('Original Image')

plt.subplot(1, 5, 2)
io.imshow(I_add)
plt.axis("off")
plt.title('Increased by 0.5')

plt.subplot(1,5, 3)
io.imshow(I_sub)
plt.axis("off")
plt.title('Deccreased by 0.4')


plt.subplot(1, 5, 4)
io.imshow(I_mul)
plt.axis("off")
plt.title('Multiplication by 2')



plt.subplot(1, 5, 5)
io.imshow(I_div)
plt.axis("off")
plt.title('Division by 3')
```

# Task 2

```python
# Import the "land.jpeg" image
I = io.imread('land.jpeg')

# normalize the image
I_norm = img_as_float(I)

# perform contrast stretching on the image

hist_org, nbins_org = exposure.histogram(I_norm, nbins=256, normalize = False)

I_stretched = np.clip(6.25*(I_norm-0.7),0,1)

hist_str, nbins_str = exposure.histogram(I_stretched, nbins=256, normalize = False)




plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.plot(nbins_org,hist_org)
plt.title('Original histogram')

plt.subplot(1, 2, 2)
plt.plot(nbins_str,hist_str)
plt.title('Stretched  histogram')



plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
io.imshow(I_norm)
plt.title('Original Image')

plt.subplot(1, 2, 2)
io.imshow(I_stretched)
plt.title('Stretched  Image')

```
